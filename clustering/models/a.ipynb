```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from pyod.models.ecod import ECOD
import prince
import plotly.express as px 
from sklearn.decomposition import PCA

# Load Data
df = pd.read_excel("./newwww.xlsx")
df_embedding = pd.read_csv("./a.csv", sep=",")

# Outlier Removal
outliers = ECOD().fit_predict(df_embedding)
df_embedding_no_out = df_embedding[outliers == 0].reset_index(drop=True)
df_no_out = df[outliers == 0].reset_index(drop=True)

# Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_embedding_no_out)

# Optimal Cluster Number (simple Silhouette-based loop)

sil_scores = []
for k in range(2, 10):
    km = KMeans(n_clusters=k, n_init=10, random_state=0)
    labels = km.fit_predict(X_scaled)
    sil_scores.append(silhouette_score(X_scaled, labels))
n_clusters = np.argmax(sil_scores) + 2  # clusters with max Silhouette
print(n_clusters)

# Final Clustering
km_final = KMeans(n_clusters=n_clusters, n_init=10, random_state=0)
clusters = km_final.fit_predict(X_scaled)

# Evaluation
print(f"Davies-Bouldin Score: {davies_bouldin_score(X_scaled, clusters):.3f}")
print(f"Calinski-Harabasz Score: {calinski_harabasz_score(X_scaled, clusters):.3f}")
print(f"Silhouette Score: {silhouette_score(X_scaled, clusters):.3f}")

# PCA for Visualization
pca = prince.PCA(n_components=3, n_iter=3, rescale_with_mean=True, rescale_with_std=True, engine='sklearn', random_state=42)
pca.fit(pd.DataFrame(X_scaled))
df_pca = pca.transform(pd.DataFrame(X_scaled))
df_pca.columns = ["comp1", "comp2", "comp3"]
df_pca["cluster"] = clusters

fig = px.scatter_3d(df_pca, x='comp1', y='comp2', z='comp3', color='cluster', title='3D PCA Clusters')
fig.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns

def optimize_dbscan_pca_practical(embeddings):
    """
    Practical optimization with reasonable parameter ranges
    Tests ~50,000 combinations (much faster than full range)
    """
    
    scaler = StandardScaler()
    embeddings_scaled = scaler.fit_transform(embeddings)
    max_features = embeddings_scaled.shape[1]
    
    # Practical ranges
    pca_range = list(range(2, min(51, max_features + 1)))  # 2 to 50 components
    eps_range = np.arange(1, 51, 1)  # 0.1 to 5.0 in 0.2 increments  
    min_samples_range = list(range(2, 31))  # 2 to 30 samples
    
    results = []
    total_combinations = len(pca_range) * len(eps_range) * len(min_samples_range)
    
    print(f"Testing {total_combinations:,} combinations with practical ranges...")
    
    pbar = tqdm(total=total_combinations, desc="Optimizing")
    
    for n_components in pca_range:
        try:
            pca = PCA(n_components=n_components, random_state=42)
            reduced = pca.fit_transform(embeddings_scaled)
            explained_variance = pca.explained_variance_ratio_.sum()
        except:
            pbar.update(len(eps_range) * len(min_samples_range))
            continue
        
        for eps in eps_range:
            for min_samples in min_samples_range:
                pbar.update(1)
                
                try:
                    db = DBSCAN(eps=float(eps), min_samples=min_samples)
                    labels = db.fit_predict(reduced)
                    
                    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
                    
                    # Skip if no valid clusters or too many clusters
                    if n_clusters < 2 or n_clusters > len(embeddings_scaled) * 0.3:
                        continue
                    
                    noise_ratio = list(labels).count(-1) / len(labels)
                    
                    # Skip if too much noise (>80%)
                    if noise_ratio > 0.8:
                        continue
                    
                    sil_score = silhouette_score(reduced, labels)
                    cal_score = calinski_harabasz_score(reduced, labels)
                    dbi_score = davies_bouldin_score(reduced, labels)
                    
                    results.append({
                        'pca_components': n_components,
                        'explained_variance': explained_variance,
                        'eps': round(eps, 1),
                        'min_samples': min_samples,
                        'n_clusters': n_clusters,
                        'noise_ratio': noise_ratio,
                        'silhouette_score': sil_score,
                        'calinski_harabasz_score': cal_score,
                        'davies_bouldin_score': dbi_score
                    })
                    
                except:
                    continue
    
    pbar.close()
    print(f"Found {len(results)} valid configurations")
    return pd.DataFrame(results)

def find_best_configs(results_df, top_n=10):
    """Find best configurations using different criteria"""
    
    if len(results_df) == 0:
        print("No valid configurations found!")
        return None
    
    # Normalize scores for composite ranking
    results_df['sil_norm'] = (results_df['silhouette_score'] - results_df['silhouette_score'].min()) / (results_df['silhouette_score'].max() - results_df['silhouette_score'].min())
    results_df['cal_norm'] = (results_df['calinski_harabasz_score'] - results_df['calinski_harabasz_score'].min()) / (results_df['calinski_harabasz_score'].max() - results_df['calinski_harabasz_score'].min())
    results_df['dbi_norm'] = 1 - (results_df['davies_bouldin_score'] - results_df['davies_bouldin_score'].min()) / (results_df['davies_bouldin_score'].max() - results_df['davies_bouldin_score'].min())
    
    # Composite score
    results_df['composite_score'] = (results_df['sil_norm'] + results_df['cal_norm'] + results_df['dbi_norm']) / 3
    
    print("🏆 TOP CONFIGURATIONS:\n")
    
    # Best by each metric
    best_sil = results_df.loc[results_df['silhouette_score'].idxmax()]
    best_cal = results_df.loc[results_df['calinski_harabasz_score'].idxmax()]
    best_dbi = results_df.loc[results_df['davies_bouldin_score'].idxmin()]
    best_composite = results_df.loc[results_df['composite_score'].idxmax()]
    
    print("📊 Best by Silhouette Score:")
    print(f"   PCA: {best_sil['pca_components']}, eps: {best_sil['eps']}, min_samples: {best_sil['min_samples']}")
    print(f"   Clusters: {best_sil['n_clusters']}, Score: {best_sil['silhouette_score']:.4f}\n")
    
    print("📊 Best by Calinski-Harabasz Score:")
    print(f"   PCA: {best_cal['pca_components']}, eps: {best_cal['eps']}, min_samples: {best_cal['min_samples']}")
    print(f"   Clusters: {best_cal['n_clusters']}, Score: {best_cal['calinski_harabasz_score']:.2f}\n")
    
    print("📊 Best by Davies-Bouldin Score (lower is better):")
    print(f"   PCA: {best_dbi['pca_components']}, eps: {best_dbi['eps']}, min_samples: {best_dbi['min_samples']}")
    print(f"   Clusters: {best_dbi['n_clusters']}, Score: {best_dbi['davies_bouldin_score']:.4f}\n")
    
    print("🎯 RECOMMENDED (Composite Score):")
    print(f"   PCA Components: {best_composite['pca_components']}")
    print(f"   DBSCAN eps: {best_composite['eps']}")
    print(f"   DBSCAN min_samples: {best_composite['min_samples']}")
    print(f"   Number of clusters: {best_composite['n_clusters']}")
    print(f"   Noise ratio: {best_composite['noise_ratio']:.3f}")
    print(f"   Silhouette: {best_composite['silhouette_score']:.4f}")
    print(f"   Calinski-Harabasz: {best_composite['calinski_harabasz_score']:.2f}")
    print(f"   Davies-Bouldin: {best_composite['davies_bouldin_score']:.4f}")
    
    return best_composite

def apply_best_params(embeddings, best_params):
    """Apply the best parameters found"""
    
    scaler = StandardScaler()
    embeddings_scaled = scaler.fit_transform(embeddings)
    
    # Apply PCA
    pca = PCA(n_components=int(best_params['pca_components']), random_state=42)
    reduced = pca.fit_transform(embeddings_scaled)
    
    # Apply DBSCAN
    db = DBSCAN(eps=best_params['eps'], min_samples=int(best_params['min_samples']))
    labels = db.fit_predict(reduced)
    
    return labels, reduced, pca

# Main execution function
def run_practical_optimization(all_embeddings):
    """Complete practical optimization pipeline"""
    
    print("🚀 Starting practical DBSCAN + PCA optimization...")
    
    # Run optimization
    results_df = optimize_dbscan_pca_practical(np.array(all_embeddings))
    
    if len(results_df) == 0:
        print("❌ No valid configurations found. Try adjusting parameter ranges.")
        return None
    
    # Save results
    results_df.to_csv('practical_optimization_results.csv', index=False)
    print("💾 Results saved to 'practical_optimization_results.csv'")
    
    # Find best parameters
    best_params = find_best_configs(results_df)
    
    if best_params is not None:
        # Apply best parameters
        labels, reduced, pca = apply_best_params(all_embeddings, best_params)
        
        print(f"\n✅ Optimization complete!")
        print(f"📋 Final Results:")
        print(f"   - Total samples: {len(all_embeddings)}")
        print(f"   - Clusters found: {len(set(labels)) - (1 if -1 in labels else 0)}")
        print(f"   - Noise points: {list(labels).count(-1)}")
        print(f"   - Explained variance: {best_params['explained_variance']:.3f}")
        
        return {
            'best_params': best_params,
            'labels': labels,
            'reduced_data': reduced,
            'pca_model': pca,
            'results_df': results_df
        }
    
    return results_df
run_practical_optimization(df_embedding)

# Usage:
"""
# Your embeddings (replace with your actual data)
a = pd.DataFrame(all_embeddings)
a.to_csv("./a.csv", index=False)

# Run practical optimization
optimization_results = run_practical_optimization(all_embeddings)

# Access results
if optimization_results:
    best_params = optimization_results['best_params']
    cluster_labels = optimization_results['labels']
    reduced_data = optimization_results['reduced_data']
    
    # Print final metrics
    print(f"Davies-Bouldin Score: {best_params['davies_bouldin_score']:.3f}")
    print(f"Calinski-Harabasz Score: {best_params['calinski_harabasz_score']:.3f}")
    print(f"Silhouette Score: {best_params['silhouette_score']:.3f}")
"""
```
