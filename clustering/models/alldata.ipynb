```python

import pandas as pd
```

```python
df28=pd.read_excel('../input_data/new/Feb28TGASP_Base.xlsx')
```

```python
df25=pd.read_excel('../input_data/new/fina_feb_25.xlsx')
```

```python
df18=pd.read_excel('../input_data/new/oct_18_min_sample21_csp_1.5_update.xlsx')
```

```python
#Colummn names
parent_Industry = 'Parent Industry'
aoi_3 = 'aoi_3'
aoi_2 = 'aoi_2'
aoi_1 = 'aoi_1'
Job_Function = 'Job_Function'
Job_Seniority = 'Job_Seniority'
```

```python
# Get count of each columns for feb28 dataset
top_2_of_cluster = (
    df28.groupby([parent_Industry])
      .size()
      .reset_index(name='count')
      .sort_values([parent_Industry, 'count'], ascending=[True, False])
      .groupby(parent_Industry)
      .head(2)
)
 
```

```python
# Get count of each columns for feb25 dataset
top_2_of_cluster = (
    df25.groupby([parent_Industry])
      .size()
      .reset_index(name='count')
      .sort_values([parent_Industry, 'count'], ascending=[True, False])
      .groupby(parent_Industry)
      .head(2)
)
 
```

```python
# Get count of each columns for oct_18 dataset
top_2_of_cluster = (
    df18.groupby([parent_Industry])
      .size()
      .reset_index(name='count')
      .sort_values([parent_Industry, 'count'], ascending=[True, False])
      .groupby(parent_Industry)
      .head(2)
)
 
```

```python
cluster_summary = (
    df28.groupby("cluster_2d_rm")
    .agg({
        "Parent Industry": lambda x: x.mode()[0] if not x.mode().empty else None,
        "Job_Seniority": lambda x: x.mode()[0] if not x.mode().empty else None,
        "Job_Function": lambda x: x.mode()[0] if not x.mode().empty else None,
        "aoi_1": lambda x: x.mode()[0] if not x.mode().empty else None,
        "aoi_2": lambda x: x.mode()[0] if not x.mode().empty else None,
        "aoi_3": lambda x: x.mode()[0] if not x.mode().empty else None,
        "email": "count"
    })
    .rename(columns={"email": "count"})
    .reset_index()
)
```

```python
top_industries = (
    df28.groupby(['cluster_2d_rm','Parent Industry'])
      .size()
      .reset_index(name='count')
      .sort_values(['cluster_2d_rm', 'count'], ascending=[True, False])
      .groupby('cluster_2d_rm')
      .head()
)
```
